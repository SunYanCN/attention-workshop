```python
#machine_vocab_file = open('machine_vocab.pickle', 'wb')
#pickle.dump(machine_vocab, machine_vocab_file)
#machine_vocab_file.close()

machine_vocab_file = open('dataset/machine_vocab.pickle', 'rb')
machine_vocab = pickle.load(machine_vocab_file)
machine_vocab_file.close()
```

```python
#inv_machine_vocab_file = open('inv_machine_vocab.pickle', 'wb')
#pickle.dump(inv_machine_vocab, inv_machine_vocab_file)
#inv_machine_vocab_file.close()

inv_machine_vocab_file = open('dataset/inv_machine_vocab.pickle', 'rb')
inv_machine_vocab = pickle.load(inv_machine_vocab_file)  # variables come out in the order you put them in
inv_machine_vocab_file.close()
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
X = []
Y = []
for i, date in enumerate(dates):
    index_x = []
    index_y = []
    for j, character in enumerate(date[0]):
        index_x.append(human_vocab.get(character))
    X.append(index_x)
    for j, character in enumerate(date[1]):
        index_y.append(machine_vocab.get(character))
    Y.append(index_y)
    
X = pad_sequences(X, maxlen=30, padding="post", value=36.)
```

```python
input_encoding = human_vocab
output_encoding = machine_vocab
input_decoding = {v: k for k, v in input_encoding.items()}
output_decoding = {v: k for k, v in output_encoding.items()}
```

```python
import numpy as np
```

```python
Xoh = np.zeros(shape=(len(X), 30, 37))
for i, data in enumerate(X): 
    for j, character in enumerate(data): 
        Xoh[i, j, character] = 1.
```

```python
Yoh = np.zeros(shape=(len(Y), 10, 11))
for i, data in enumerate(Y): 
    for j, character in enumerate(data): 
        Yoh[i, j, character] = 1.
```

```python
index = 0
print("Source date:", dates[index][0])
print("Target date:", dates[index][1])
print()
print("Source after preprocessing (indices):", X[index])
print("Target after preprocessing (indices):", Y[index])
print()
print("Source after preprocessing (one-hot):", Xoh[index])
print("Target after preprocessing (one-hot):", Yoh[index])
```
