```python
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
dates_file = open('dataset/dates.pickle', 'rb')
dates = pickle.load(dates_file)
dates_file.close()
```

```python
X = []
Y = []
for i, date in enumerate(dates):
    X.append(date[0])
    Y.append(date[1])
```

```python
index_x = []
index_y = []
for i, date in enumerate(dates):
    for j, character in enumerate(date[0]):
        index_x.append(character)
    for j, character in enumerate(date[1]):
        index_y.append(character)
```

```python
human_vocab_vectorizer = CountVectorizer(analyzer='char')
human_vocab = human_vocab_vectorizer.fit(X)
machine_vocab_vectorizer = CountVectorizer(analyzer='char')
machine_vocab = machine_vocab_vectorizer.fit(Y)
```

```python
print(len(human_vocab.vocabulary_))
print(len(machine_vocab.vocabulary_))
```
